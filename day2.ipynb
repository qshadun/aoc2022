```python
d = {'X': 'A', 'Y': 'B', 'Z': 'C'}
s1 = {'A': 1, 'B': 2, 'C': 3}
```

```python
def comp(a, b):
    b = d[b]
    if a == b:
        return 3
    if a == 'A':
        if b == 'B':
            return 6
        else:
            return 0
    elif a == 'B':
        if b == 'A':
            return 0
        else:
            return 6
    else:
        if b == 'A':
            return 6
        else:
            return 0
```

```python
test = """A Y
B X
C Z"""

ans = 0
for line in test.split('\n'):
    a, b = line.split(' ')
    print(a, b, comp(a, b), s1[d[b]])
    ans += comp(a, b) + s1[d[b]]
ans
```

```python
def choose(a, b):
    if b == 'Y':
        return a
    if b == 'Z':
        if a == 'A':
            return 'B'
        elif a == 'B':
            return 'C'
        else:
            return 'A'
    else:
        if a =='A':
            return 'C'
        elif a == 'B':
            return 'A'
        else:
            return 'B'
        
s2 = {'X': 0, 'Y': 3, 'Z': 6}
```

```python
def score2(a, b):
    ans = s2[b]
    ans += s1[choose(a, b)]
    return ans
    
```
